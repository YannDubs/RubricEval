```python
from helm_instruct import Instructionator, Rubricator, Completor, Evaluator
from alpaca_eval import utils as ae_utils
import ast
import pandas as pd
pd.set_option('display.max_colwidth', None)
```

```python
n_max_examples =20
```

```python
instructionator = Instructionator()
instructions = instructionator.generate_n_instructions_per_categories(max(2//11,1))
```

```python
for instruction in instructions[:n_max_examples]:
    print("Category: ", instruction["category"])
    print("Prompt: ", instruction["prompt"])
    print("---------------------\n\n\n")
```

```python
rubricator = Rubricator()
rubrics = rubricator(instructions[:n_max_examples])
```

```python
df_rubrics = rubricator.make_df_rubrics(rubrics)
for i in range(min(len(df_rubrics), n_max_examples)):
    
    print("Example: ", i)
    print("Category: ", df_rubrics.loc[i, "category"])
    print("\nPrompt: ", df_rubrics.loc[i, "final_prompt"])
    display(pd.DataFrame(df_rubrics.loc[i, "detailed_analytic_rubric"]).T)
    print("---------------------\n\n\n")
```

```python
claude = Completor(annotators_config="claude-2")
claude_completions = claude(df_rubrics[:n_max_examples])
```

```python
for completion in claude_completions[:n_max_examples]:
    print("Category: ", completion["category"])
    print("\n#Prompt:\n", completion["final_prompt"])
    print("\n#Output:\n", completion["output"])
    print("---------------------\n\n\n")
```

```python
chatgpt = Completor(annotators_config="chatgpt")
chatgpt_completions = chatgpt(df_rubrics[:n_max_examples])
```

```python
for completion in chatgpt_completions[:n_max_examples]:
    print("Category: ", completion["category"])
    print("\n#Prompt:\n", completion["final_prompt"])
    print("\n#Output:\n", completion["output"])
    print("---------------------\n\n\n")
```

```python
cohere = Completor(annotators_config="cohere-chat")
cohere_completions = cohere(df_rubrics[:n_max_examples])
```

```python
for completion in cohere_completions[:n_max_examples]:
    print("Category: ", completion["category"])
    print("\n#Prompt:\n", completion["final_prompt"])
    print("\n#Output:\n", completion["output"])
    print("---------------------\n\n\n")
```

```python
evaluator = Evaluator()
chatgpt_scores = evaluator(chatgpt_completions)
```

```python
import numpy as np
def get_mean_value(x):
    try:
        seq = list(ast.literal_eval(x).values())
    except :
        seq = [np.nan]
    return np.mean(seq)
        
metrics = ae_utils.convert_to_dataframe(chatgpt_scores)["criteria_scores"].apply(get_mean_value)
print("chatgpt", metrics.mean())
```

```python
claude_scores = evaluator(claude_completions)
metrics = ae_utils.convert_to_dataframe(claude_scores)["criteria_scores"].apply(get_mean_value)
print("claude", metrics.mean())
```

```python
cohere_scores = evaluator(cohere_completions)
metrics = ae_utils.convert_to_dataframe(cohere_scores)["criteria_scores"].apply(get_mean_value)
print("cohere", metrics.mean())
```
